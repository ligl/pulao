```python
import os
import pandas as pd

def parse_tick_datetime(s):
    try:
        return pd.to_datetime(s, format="%Y%m%d%H%M%S.%f")
    except ValueError:
        return pd.to_datetime(s, format="%Y%m%d%H%M%S")

csv_path = os.path.abspath("dataset/I2601.XDCE_tick.csv")
df = pd.read_csv(csv_path)
df = df.head(100)
df["datetime"] = df["datetime"].astype(str).apply(lambda x: x if "." in x else x + ".000") # 统一数据格式
df["datetime"] = pd.to_datetime(df["datetime"], format="%Y%m%d%H%M%S.%f")

#df["datetime"] = df["datetime"].apply(parse_tick_datetime)

df
```
